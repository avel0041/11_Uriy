Запишите слово, которое стоит под номером 75.

```python
from itertools import*
n=0
for x in product(sorted('влту'), repeat=4):
    a = ''.join(x)
    n+=1
    if n ==75:
        print(a)
        break

print(list(product(sorted('влту'), repeat=4))[74])
```

Шифр кодового замка представляет собой последовательность из пяти символов, каждый из которых является цифрой от 1 до 4.
